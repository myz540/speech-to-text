```python
import speech_recognition as sr 
import os 
from pydub import AudioSegment
from pydub.silence import split_on_silence
from pynput import keyboard
import pyaudio
```

```python
r = sr.Recognizer()
with sr.Microphone() as source:
    # read the audio data from the default microphone
    audio_data = r.record(source, duration=5)
    print("Recognizing...")
    # convert speech to text
    text = r.recognize_google(audio_data)
    print(text)
```

```python
class AudioRecorder:
    
    def __init__(self):
        self.recognizer = sr.Recognizer()
        self.microphone = sr.Microphone()
        
        self.is_recording = False

        with self.microphone as source:
            print('calibrating...')
            self.recognizer.adjust_for_ambient_noise(source)
        
        self.keyboard_controller = keyboard.Controller()
        self.keyboard_listener = keyboard.Listener(on_press=self.on_press, on_release=self.on_release)
        self.keyboard_listener.start()
        
    
    def callback(self, recognizer, audio):
        print("in callback")
        # received audio data, now we'll recognize it using Google Speech Recognition
        try:
            # for testing purposes, we're just using the default API key
            # to use another API key, use `r.recognize_google(audio, key="GOOGLE_SPEECH_RECOGNITION_API_KEY")`
            # instead of `r.recognize_google(audio)`
            data = recognizer.recognize_google(audio)
            chars = list(data)
            for c in chars:
                self.keyboard_controller.press(c)
            print("Google Speech Recognition thinks you said " + data)
        except sr.UnknownValueError:
            print("Google Speech Recognition could not understand audio")
        except sr.RequestError as e:
            print("Could not request results from Google Speech Recognition service; {0}".format(e))
        
        
    def on_press(self, key):
        try:
            print('alphanumeric key {0} pressed'.format(
                key.char))
        except AttributeError:
            print('special key {0} pressed'.format(
                key))
        if key == keyboard.Key.cmd or key == keyboard.Key.cmd_r:
            if not self.is_recording:
                self.is_recording = True
                print('starting audio listener')
                self.stop_listening = self.recognizer.listen_in_background(self.microphone, self.callback)
            else:
                self.is_recording = False
                print('stopping audio listener')
                self.stop_listening(wait_for_stop=False)
                
    
    def on_release(self, key):
        print('{0} released'.format(
            key))
        if key == keyboard.Key.cmd:
            pass
        if key == keyboard.Key.esc:
            print("stopping keyboard listener")
            return False
        

        
```
